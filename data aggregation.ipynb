### weather

```python
stations = pd.read_csv("stations.csv")
```

```python
stations.rename(columns={"0": "IIiii", '4': 'height', "5": "country"}, inplace=True)
```

```python
stations.head()
```

```python
stations.drop(stations.columns.difference(['IIiii', 'height', 'country']), 1, inplace=True)
```

```python
stations.dropna(inplace=True)
```

```python
stations['height'] = stations.apply(lambda row: " ".join(row['height'].split()), axis=1)
stations['country'] = stations.apply(lambda row: " ".join(row['country'].split()), axis=1)
```

```python
stations['country'][0]
```

```python
stations['IIiii'] = pd.to_numeric(stations['IIiii'])
```

```python
stations['height'] = pd.to_numeric(stations['height'])
```

```python
stations = stations[stations['height'] < 500]
```

```python
stations[stations['country'] == 'Russian Federation']
```

#### weather preprocess

```python
we = pd.read_csv("weather_data.csv")
we.drop(we.columns.difference(['year', 'month', 'IIiii', 'sn', 'T', 'S1', 'R1', 'R01', 'f20']), 1, inplace=True)
we.head()
```

```python
we.describe()
```

```python
we['T'] = we.apply(lambda row : row['T'] / 10 if row['sn'] == 0 else -row['T'] / 10, axis=1)
```

```python
we = we[(6 <= we['month']) & (we['month'] <= 8)]
```

```python
we.describe()
```

#### merging

```python
wem = pd.merge(we, stations, how="inner", on=["IIiii"])
```

```python
useful = wem.groupby(['country']).mean()
```

```python
useful.drop(wem.columns.difference(['country', 'T', 'S1', 'R1', 'f20']), 1, inplace=True)
```

```python
useful.rename(columns={'T': 'temp', 'S1': 'sunshine', 'R1': 'precipitation', 'f20': 'windy days'}, inplace=True)
```

### scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
m = pd.read_csv("full_data.csv")
```

```python
m.info()
```

```python
m.head()
```

```python
def percents(v):
    if type(v) == type(.0):
        return np.nan
    else:
        return v[:-1]

m['index_religion'] = pd.to_numeric(m.apply(lambda row: percents(row['index_religion']), axis=1))
```

```python
m.info()
```
